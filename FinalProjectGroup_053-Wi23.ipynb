For this method, we first typecast every input into integer, then use a for loop to find a match in the data frame new_cases we just created. then set the output to be the covid level that corresponds to that new case number.

```python
def add_level(string):
    string = int(string)
    for i in range(new_cases.shape[0]):
        if new_cases.iloc[i,-2] == string:
            output = new_cases.iloc[i,-1]
            break
    return output
```

# Conclusion & Discussion

When we started our group project, we were interested in finding a direct relationship between the covid severity level and mental health issues in the United States. We believed that with the surge of Covid-19, people of different categories in the society were all affected mentally, and we would like to locate one specific group that was influenced the most by Covid-19. We collect data on the percentage of people taking Prescription Medication for Mental Health for different categories such as race, gender, education level, sex, and disability status, covid cases for each day. In this way, we can quantify covid severity level by number of new cases and mental health issues by percentage of people taking Prescription Medication for Mental Health. The initial idea was to analyze how mental health as represented by the ‘Value’ column in the data frame in a survey covering distinct social communities changes with the change of weekly new cases increase. Based on the correlation coefficient we found for each category, we can conclude that the mental states of white women over 80 with a high school diploma who are not disabled fluctuate the most with Covid-19 severity. Because all correlation coefficients are positive, we know that as Covid severity level increases, this particular group of people tend to have more mental health issues.


There are several limits in our research, one apparent drawback is that correlation coefficient doesn’t mean causation. So, we can’t conclude that covid severity level is the reason for the increase in mental health, we can only state that there’s a positive linear relationship between covid severity level and mental health issue. Also, the correlation coefficients for that specific group are all between 0.3 and 0.5, which means that the linear relationship between the covid severity level and mental health issues were not as strong as we thought. So the group we find may not be one hundred percent accurate. Also, the data we found did not do a great job with a strict control variable process. What we tried is to control the covid variable as we can to ensure the groups are in the same time slots. However, for people of different ages, for example, their race and education status were not given and hence may differ. Therefore, our work is a general analysis of those social factors that may still incorporate a margin of error.
